# Uppgift 4

```python
class Square:
    def __init__(self, side_length):
        self.side_length = side_length

    def perimeter(self):
        return self.side_length * 4

    def area(self):
        return self.side_length ** 2

my_square = Square(8)

print(my_square.perimeter())
print(my_square.area())
```

# Uppgift 6

```python
class BankAccount:
    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"{amount} kronor sattes in på kontot.")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Balance too low")
        else:
            self.balance -= amount
            print(f"{amount} kronor togs ut från kontot.")


my_account = BankAccount("Emma", 500)

print("Kontoinnehavare:", my_account.account_holder)
print("Saldo:", my_account.balance)

my_account.deposit(200)

my_account.withdraw(100)

my_account.withdraw(700)
print("Nytt saldo:", my_account.balance)

```

En klass är som en mall man kan använda för att skapa objekt med olika så kallade attributera, till exempel kan man ha en klass som representerar en människa. Då kan attributen vara hårfärg, längd och vikt. När vi sedan skapar en instans av klassen, alltså en specifik person (ett objekt), så använder vi mallen och fyller i de värden vi vill ha. Till exempel: hårfärg = svart, längd = 160 cm, vikt = 70 kg. På så sätt kan vi skapa flera olika människor (instanser), alla baserade på samma mall (klass), men med olika attribut.
