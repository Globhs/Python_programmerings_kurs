# Uppgift 10

for i in range(100):
if i in (5, 7, 8):
continue
if i > 10:
break
print(i)
else:
print('We are done!')


Loopen kommer att printa ut i tills att den kommer till 5, 7, 8 för dem kommer den hoppa över då det inte finns någor print(i) där, men när den kommer till 10 så ska den breaka. samt printa "we are done!".


```python
for i in range(100):
    if i in (5, 7, 8):
        continue
    if i > 10:
        break
    print(i)
else:
    print('We are done!')
```
