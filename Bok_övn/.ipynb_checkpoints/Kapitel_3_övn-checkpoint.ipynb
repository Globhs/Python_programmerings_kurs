# Uppgift 10

Skillnaden på dessa är att den första exekverar alla if satser, då alla if satser är oberoende av varandra.
Medans den andra koden använder elif vilket betyder gör att if satserna är beroende av varandra och då bara printa ut en utav dem.

```python
number = 8

if number < 5:
    print('Less than 5')
if number < 10:
    print('Less than 10')
if number < 100:
    print('Less than 100')

print("------------------------------------------------------")

if number < 5:
    print('Less than 5')
elif number < 10:
    print('Less than 10')
elif number < 100:
    print('Less than 100')
```
