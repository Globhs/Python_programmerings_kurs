```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
hr_data = pd.read_excel("Hr_Employee_Data.xlsx")
```

```python
hr_data.head()
```

```python
hr_data.info()
```

```python
time_spend_left = hr_data.groupby("left")["time_spend_company"].mean()
print(time_spend_left)
```

```python
time_spend_left.plot(kind="bar", color="skyblue")
plt.title("Average Time Spent By Employee, left")
plt.xlabel("Left (0 = Stayed, 1 = Left)")
plt.ylabel("Average Time Spent")
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
